```python
import pandas
```

```python
df = pandas.read_csv('ages_ger.csv')
```

```python
df
```

```python
import matplotlib
font = {'family' : 'Consolas',
        'weight' : 'bold',
        'size'   : 60}

matplotlib.rc('font', **font)

import matplotlib.pyplot as plt
fig, ax = plt.subplots(facecolor='black', figsize=(25, 25))
ax.tick_params(colors='cyan')
df.plot.area(x='age_id', y='GER',color='cyan', alpha=0.2, xticks=[], ax=ax)
age_id_map = {'6-10': 0, '11-13': 1, '14-15': 2, '16-17': 3, '18-23': 4}
df['age_id'] = df.Age.map(age_id_map)
df.plot.scatter(x='age_id', y='GER', c='cyan', ax=ax, xticks=[],s=100)
ax.set_xlabel(None)
ax.set_ylabel('General Enrollment Rate')
ax.get_legend().remove()
ax.set_facecolor('black')

plt.tight_layout(h_pad=20)
```
